```python
import cv2
```

```python
img = cv2.imread('image.webp')
```

The image is in BGR format

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

**Converting BGR to RGB**

```python
rbg =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rbg)
plt.title("Original Image")
plt.axis("off")
```

```python
print(img.shape)
#(height, Width, Channel)
```

```python
print(img.size)
#Value in pixel
```

**Resizing the Image**

```python
resize = cv2.resize(rbg, (224,224))
plt.imshow(resize)
print(resize.shape)
```

```python
import numpy as np
```

**Croping the Image**

```python
crop = rbg[100:500, 100:400]
plt.imshow(crop)
```

**Fliping the image Horizonatly and Vertically**

```python
flip_horizontal = cv2.flip(rbg, 1)
plt.imshow(flip_horizontal)
```

```python
flip_vertical = cv2.flip(rbg, 0)
plt.imshow(flip_vertical)
```
